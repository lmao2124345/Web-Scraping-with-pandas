# Read .csv from Multiple URLs with pandas

-https://www.football-data.co.uk/mmz4281/2526/E0.csv

-https://www.football-data.co.uk/mmz4281/2526/E1.csv

Link:root+season+leauge

```python
#link structure
"https://www.football-data.co.uk/mmz4281/"+"2526"+"/"+"E0"+".csv"

```

```python
# Create root variable 
root ="https://www.football-data.co.uk/mmz4281/"
```

## Multiple leauges 
-https://www.football-data.co.uk/mmz4281/2526/E0.csv

-https://www.football-data.co.uk/mmz4281/2526/E1.csv

-https://www.football-data.co.uk/mmz4281/2526/E2.csv

-https://www.football-data.co.uk/mmz4281/2526/E3.csv

-https://www.football-data.co.uk/mmz4281/2526/E4.csv



```python
# creating list of leauges 
leauges=['E0','E1','E2','E3','EC']
frames=[]

# looping thirough leauges, read multiple csv and append it into a list
for leauge in leauges:
    df=pd.read_csv(root +'2526' + '/' + leauge+ '.csv')
    frames.append(df)
```

```python
# length of frames 
len(frames)
```

```python
#show 1st,2nd and 3rd elemnt
frames[0]
```

## Multiple seasons


```python
# creating list of leauges 
leauges=['E0','E2','E3']
frames=[]

# loop through range of seasons and insert new column 
for leauge in leauges:
    for season in range(20,26):
        df=pd.read_csv(root + str(season) + str(season+1) + '/' + leauge+ '.csv')
        df.insert(1, 'season',season)
        frames.append(df)
```

```python
#total frames(6 seasons x 3 leauges)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[17]
```

# Organize all the data in a dictionary

```python
# create a dictionary with original names of leauge as key 
dict_countries={'Spanish La Liga':'SP1','Spanish Segunda Division':'SP2','German Bundesliga':'D1','English Premier Leauge':'E0','English Leauge 1':'E2','English leauge 2':'E3'}
```

```python
for leauge in dict_countries:
    print(dict_countries[leauge])
```

```python
dict_historical_data = {}
# creating list of leauges
#leauges['E0','E2','E3']

# loop through range of seasons and insert new column 
for leauge in dict_countries:
    frames=[]
    for season in range(20,26):
        df=pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[leauge]+ '.csv')
        df.insert(1, 'season',season)
        frames.append(df)
    df_concat = pd.concat(frames)    
    dict_historical_data[leauge] = df_concat
```
